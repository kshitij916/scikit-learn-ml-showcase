# Titanic Dataset — Exploratory Data Analysis

This notebook performs exploratory data analysis on the Titanic dataset.

Goals of this analysis:

- Understand dataset structure
- Identify missing values
- Analyze feature distributions
- Explore relationships between variables
- Generate insights useful for machine learning models

```python
# Import Libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# Load Dataset
df = pd.read_csv("../data/titanic.csv")

df.head()
```

```python
# Dataset Overview
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
# Summary Statistics
df.describe()
```

```python
# Missing Values Analysis
df.isnull().sum()
```

```python
# Visualization:
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Target Variable Distribution
# Target: Survived
sns.countplot(x="Survived", data=df)

plt.title("Survival Distribution")
plt.show()
```

```python
# Survival by Gender
sns.countplot(x="Sex", hue="Survived", data=df)

plt.title("Survival by Gender")
plt.show()
```

```python
# Survival by Passenger Class
sns.countplot(x="Pclass", hue="Survived", data=df)

plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(8,5))

sns.histplot(df["Age"].dropna(), bins=30)

plt.title("Age Distribution")
plt.show()
```

```python
# Fare Distribution
plt.figure(figsize=(8,5))

sns.histplot(df["Fare"], bins=40)

plt.title("Fare Distribution")
plt.show()
```

```python
# Correlation Analysis
# Select numeric columns:
numeric_df = df.select_dtypes(include=["float64","int64"])

plt.figure(figsize=(10,6))

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Boxplots (Outlier Detection)
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Fare"])

plt.title("Fare Outliers")
plt.show()
```

## Key Insights from EDA

1. Female passengers had a much higher survival rate.
2. First-class passengers were more likely to survive.
3. The Age feature contains missing values that must be handled during preprocessing.
4. The Cabin feature has too many missing values and may be removed.
5. Fare distribution is highly skewed and contains outliers.
6. Passenger class and fare appear to be important predictors.
